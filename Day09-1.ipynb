```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('penguins')
#df = sns.load_dataset('diamonds')
df.head()
```

```python
df['species'].unique()
df['species'].value_counts()

sns.displot(data=df, x='species')
sns.displot(data=df, x='island')
```

```python
fig = plt.figure(figsize=[12,4], facecolor='lavender')
axe_left = fig.add_subplot(1,2,1)
axe_left.set_title('Species')
sns.histplot(data=df, x='species', ax=axe_left, facecolor='red')

axe_right = fig.add_subplot(1,2,2)
axe_right.set_title('Island')
sns.histplot(data=df, x='island', ax=axe_right)
```

```python
fig = plt.figure(figsize=[20,4], facecolor='lavender')
fig.suptitle('Bill_length_mm')

axe_left = fig.add_subplot(1,3,1)
axe_left.set_title('By Species')
sns.histplot(data=df, x='bill_length_mm', ax=axe_left, hue='species', palette=['r','g','b'])

axe_middle = fig.add_subplot(1,3,2)
axe_middle.set_title('By Island')
sns.histplot(data=df, x='bill_length_mm', ax=axe_middle, hue='island', multiple='dodge', kde=True)

axe_right = fig.add_subplot(1,3,3)
axe_right.set_title('By Sex')
sns.ecdfplot(data=df, x='bill_length_mm', ax=axe_right, hue='sex')
```
