```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
print(df.head())
```

```python
sns.pairplot(df)
```

```python
fig = plt.figure(figsize=[12,4])
a = fig.add_subplot(1,2,1)
sns.scatterplot(data=df, x='day', y='total_bill', ax=a, hue='sex')

b = fig.add_subplot(1,2,2)
sns.histplot(data=df, x='day', y='total_bill', ax=b, cbar=True)
```

```python
fig = plt.figure(figsize=[12,4])
a = fig.add_subplot(1,2,1)
#sns.stripplot(data=df, x='day', y='total_bill', ax=a, jitter=0.2)
sns.stripplot(data=df, x='day', y='total_bill', ax=a, jitter=0.1, hue='sex', legend=False, dodge=True)

b = fig.add_subplot(1,2,2)
sns.swarmplot(data=df, x='day', y='total_bill', ax=b, hue='sex', dodge=True)
```

```python
fig = plt.figure(figsize=[20,4])
a = fig.add_subplot(1,3,1)
sns.boxplot(data=df, x='day', y='total_bill', ax=a, hue='sex')

b = fig.add_subplot(1,3,2)
sns.violinplot(data=df, x='day', y='total_bill', ax=b, bw=0.5, cut=0)
#sns.catplot(data=df, x='day', y='total_bill', kind='violin', hue='sex', split=True, inner='stick')

c = fig.add_subplot(1,3,3)
sns.boxenplot(data=df, x='day', y='total_bill', ax=c)
```

```python
fig = plt.figure(figsize=[20,4])
a = fig.add_subplot(1,3,1)
a.set_ylim(0,25)
sns.barplot(data=df, x='day', y='total_bill', ax=a, hue='sex')

b = fig.add_subplot(1,3,2)
b.set_ylim(0,25)
sns.pointplot(data=df, x='day', y='total_bill', ax=b, hue='sex', dodge=True)

c = fig.add_subplot(1,3,3)
sns.countplot(data=df, x='day', ax=c)
```

```python
fig = plt.figure(figsize=[12,4])
a = fig.add_subplot(1,2,1)
sns.countplot(data=df, x='day', ax=a)

b = fig.add_subplot(1,2,2)
sns.histplot(data=df, x='day', shrink=0.7)
```

```python
df1 = sns.load_dataset('flights')
print(df1)
df2 = df1.pivot(index='month', columns='year', values='passengers')
print(df2)
#sns.heatmap(df2, annot=True, fmt='d')
sns.heatmap(df2)
```

```python
a = df[['day','time','tip']].dropna()
#print(a)
b = a.groupby(['day','time']).sum()
b = b.reset_index()
print(b)
b = b.pivot('time','day','tip')
print(b)
sns.heatmap(b, annot=True, fmt='g')
```

```python
fig = plt.figure(figsize=[12,4])
a = fig.add_subplot(1,2,1)
sns.regplot(data=df, x='total_bill', y='tip')

b = fig.add_subplot(1,2,2)
sns.residplot(data=df, x='total_bill', y='tip')

sns.lmplot(data=df, x='total_bill', y='tip')   #fig level
```
