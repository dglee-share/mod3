```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_tips = sns.load_dataset('tips')
print(df_tips)
print(type(df_tips))
```

```python
df_tips['tip_percent'] = (df_tips['tip']*100)/df_tips['total_bill']
print(df_tips)
```

```python
sns.relplot(data=df_tips, x='total_bill', y='tip_percent', hue='time')
```

```python
import numpy.random as npr

df1 = pd.DataFrame(npr.randn(1000,1).cumsum(), columns=['A'])
#print(df1)
sns.relplot(data=df1, x=df1.index, y='A', kind='line', height=3, aspect=1.5)
```

```python
x = sns.relplot(data=df1, x=df1.index, y='A', kind='line', height=3, aspect=1.0)
print(type(x.figure))
f = x.figure
f.set_facecolor('red')
a = f.axes
print(type(a[0]))
a[0].set_ylim(-30, 30)
```

```python
#axe level
fig1 = plt.figure(tight_layout=True)
a = fig1.add_subplot(2,1,1)
b = fig1.add_subplot(2,2,3)
c = fig1.add_subplot(2,2,4)
sns.scatterplot(data=df1, x=df1.index, y='A', ax=b)
sns.lineplot(data=df1, x=df1.index, y='A', ax=c)
```
